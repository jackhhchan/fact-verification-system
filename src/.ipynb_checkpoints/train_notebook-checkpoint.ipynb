```python
import tensorflow as tf

file_pattern = "../dataset/tfrecords/train_512.tfrecord"
files = tf.data.Dataset.list_files(file_pattern)

AUTOTUNE  = tf.data.experimental.AUTOTUNE
ds = files.interleave(lambda x:
                tf.data.TFRecordDataset(x),
                cycle_length=4,
                num_parallel_calls=AUTOTUNE)
print("Extracted all TFRecords with pattern {}".format(file_pattern))
```

```python
x = next(iter(ds))
x
```

```python
for raw_record in ds.take(1):
    example = tf.train.Example()
    example.ParseFromString(raw_record.numpy())
    print(example)
```

```python
feature_description = {
    'input_word_ids': tf.io.FixedLenFeature([512], tf.int64),
    'input_mask': tf.io.FixedLenFeature([512], tf.int64),
    'segment_ids': tf.io.FixedLenFeature([512], tf.int64),
    'target': tf.io.FixedLenFeature([1], tf.int64)
}

def parse_and_transform(serialized)->tuple:
    example = tf.io.parse_single_example(serialized, feature_description)
    # transform
    target = example.pop('target')
    target = tf.reshape(target, ())
    
    embeddings_dict = example
    target_dict = {'target': target}

    return (embeddings_dict, target_dict)

transformed = parse_and_transform(x)
transformed
# parsed_ds = ds.map(parse_and_transform)
```
